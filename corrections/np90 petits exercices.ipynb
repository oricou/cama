### Norme d'un vecteur

Retourner la norme euclidienne d'un vecteur sans utiliser la fonction `norm`.

2 lignes

```python
def norm(v):
    return np.sqrt(v @ v)
```

### Sous-matrice

Extraire la sous-matrice de taille (n-2)x(m-2) qui retire les bords de la matrice A de taille nxm.

2 lignes

```python
def submat(A):
    return A[1:-1, 1:-1]
```

### Vecteur aléatoire

Créer une fonction qui prend une taille n et retourne un vecteur  d'entiers aléatoire de taille n compris entre -a et +a et centré le plus possible en 0.

3 lignes

```python
def rand_vec(n, a):
    res =  np.random.randint(-a, a, size=n)
    return res - round(res.mean())
```

### Trace

Retourner la trace d'une matrice sans utiliser la fonction `trace` (à vous de chercher les éléments qui vous manque).

2 lignes

```python
def trace(A):
    return A.diagonal().sum()
```

Retourner l'indice de la colonne d'une matrice ayant la plus petite moyenne.

```python
def min_col_mean(A):
    return A.mean(axis=0).argmin()
```

### ChessSum

On regarde la matrice comme un échequier avec des cases blanches et des cases noires qui alternent. La case en [0,0] est noire. 
Calculer la somme des valeurs sur les cases blanches.

2 lignes

```python
def chess_sum(A):
    return A[1::2,::2].sum() + A[::2, 1::2].sum()
```

### 2 minimums

Retourner les 2 plus petites valeurs d'une matrice avec une méthode de complexité linéaire. Attention, si la matrice a sa valeur minimal en double, la réponse est deux fois cette valeur minimale.

6 lignes

```python
def mins(A):
    i1 = A.argmin()       # 1D indice
    res1 = A.flat[i1]
    A.flat[i1] = A.max()  # change A
    return res1, A.min()
```

```python
def mins(A):
    res1 = A.min()
    if (A == res1).sum() > 1:  # un filtre prend de l'espace mémoire
        return res1, res1
    else:
        return res1, A[A > res1].min()
```

```python
def mins(A):               # le meilleur mais 8 lignes
    i = A.argmin()
    if i ==0:
        return A.flat[i], A.flat[1:].min()
    elif i==len(A):
        return A.flat[i], A.flat[:-1].min()
    else:
        return A.flat[i], min(A.flat[:i].min(), A.flat[i+1:].min())
```

### Lignes dans l'ordre

Ranger les lignes d'une matrice dans l'ordre croissant de leur moyenne.

2 lignes

```python
def sort_lines(A):
    return A[A.sum(axis=1).argsort()]  # sum est plus rapide que mean et donne le même résultat
```

### Valeurs uniques

Donner la liste des valeurs uniques (qui n'apparaissent qu'une fois) dans un tableau numpy (la fonction `unique` sera utile).

3 lignes

```python
def val_unique(A):
    val, nb = np.unique(A, return_counts=True)
    return val[nb == 1]
```

### Tenseur magique

Constuire un tenseur 3D de taille nxnxn en utilisant n fois les n² premiers entiers et de sorte que la sommme des éléments de chaque plan (coupe) est toujours la même.

3 lignes

```python
def mk_magic_tensor(n):
    A = np.arange(n*n).reshape([n,n])
    return np.stack([np.roll(np.roll(A, i, axis=0), i, axis=1) for i in range(n)])
```

5 lignes (2 lignes avec la fonction `take`)

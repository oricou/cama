### Exercice 20.1

Comment se fait-il que la courbe du résidu soit la même que celle de l'erreur relative ?

**Réponse**

L'erreur relative est

$$
\begin{align*}
{\bf x}^{k+1} - {\bf x}^k &=  w \, M^{-1} \, (N\; {\bf x}^k + {\bf b}) + (1-w) \, {\bf x}^k - {\bf x}^k \\
                           &=  w \, M^{-1} \, (N\; {\bf x}^k + {\bf b}) -w \, {\bf x}^k \\
                           &=  w \, M^{-1} \, ((N - M)\; {\bf x}^k + {\bf b})\\
                           &=  w \, M^{-1} \, (-A\; {\bf x}^k + {\bf b})\\
\end{align*}
$$
et donc
$$
||{\bf x}^{k+1} - {\bf x}^k|| =  |w| \; ||M^{-1}|| \; ||-A\; {\bf x}^k + {\bf b}||
$$

On retrouve le résidu $||A \, {\bf x} - {\bf b}||$ ce qui explique que les 2 courbes soient égales à un coefficient multiplicatif près.
